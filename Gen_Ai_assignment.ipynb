```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
# Load GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
model.eval()
```

```python
# Define Function to Generate Story
def generate_story(prompt, max_length=100, temperature=1.0, top_k=50):
    # Encode the input prompt
    inputs = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text
    outputs = model.generate(
        inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        num_return_sequences=1,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode output
    story = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return story
```

```python
story = generate_story("The spaceship landed on Mars and")
print(story)
```

```python
# Prompt 1
prompt1 = "Once upon a time in a hidden forest,"
story1 = generate_story(prompt1, max_length=80, temperature=0.8, top_k=40)
print("Prompt 1 Output:")
print(story1)
```

```python
# Prompt 2
prompt2 = "The robot opened its eyes and realized it was alone."
story2 = generate_story(prompt2, max_length=80, temperature=1.2, top_k=50)
print("prompt 2 output:")
print (story2)
```

```python
# Prompt 3
prompt3 = "In the year 3025, humanity discovered a new planet."
story3 = generate_story(prompt3, max_length=80, temperature=0.9, top_k=30)
print("Prompt 3 Output:")
print(story3)
```
